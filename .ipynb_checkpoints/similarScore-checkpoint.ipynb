```python
from __future__ import unicode_literals
from selenium import webdriver
# import selenium.webdriver.support.ui as ui
from selenium.webdriver.common.keys import Keys
import time
import sys
import requests
from zipfile import ZipFile
import os
from bs4 import BeautifulSoup
import pandas as pd
import requests
import re

driver_link = 'https://chromedriver.storage.googleapis.com/2.44/chromedriver_win32.zip'

r = requests.get(driver_link)  
with open('chromeDriver.zip', 'wb') as f:
    f.write(r.content)
with ZipFile('chromeDriver.zip','r') as zip:
    zip.extractall()
os.remove("chromeDriver.zip")
driver = webdriver.Chrome("chromedriver.exe")
url = 'https://www.christies.com/impressionist-and-modern-art-27255.aspx?lid=1&dt=100420180204&saletitle='
driver.get(url)
time.sleep(10)
close_sign_up_xpath = '//*[@id="close_signup"]'
load_btn_xpath = '//*[@id="loadAllUpcomingPast"]'
links_xpath = '//*[@id="ResultContainer"]/li/div/div[1]/a[1]'
try:
    close_signup_element  = driver.find_element_by_id(close_sign_up_xpath)
    close_signup.click()
except:
    pass
    
html = driver.find_element_by_tag_name('html')
html.send_keys(Keys.END)
load_all_button_element = driver.find_element_by_xpath(load_btn_xpath)
load_all_button_element.click()
time.sleep(5)
links_element = driver.find_elements_by_xpath(links_xpath)
links_list = [i.get_attribute('href') for i in links_element]
```

```python
for each in links_list:
    print(each + "\n\n")
```

```python

def cleanText(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)

def scrapeLots(url):
	wt_ht = []
	pr = requests.get(url,timeout=5)
	pc = BeautifulSoup(pr.content, 'html.parser')

	# scraping values from webpage
	art_name = pc.find(class_="itemName")
	art_name = cleanText(str(art_name))
	lot_number = pc.find("span", id="main_center_0_lblLotNumber")
	lot_number = cleanText(str(lot_number))
	artist_name = pc.find("span", {"id": "main_center_0_lblLotPrimaryTitle"})
	artist_name = cleanText(str(artist_name))
	price = pc.find("p", id="main_center_0_lblPriceRealizedPrimary")
	price = cleanText(str(price))
	estimate = pc.find("span", id = "main_center_0_lblPriceEstimatedPrimary")
	estimate = cleanText(str(estimate))
	image = pc.find("source", media="(max-width: 1152px)")
	details = pc.find("span", id="main_center_0_lblLotDescription")
	provenance = pc.find("p", id="main_center_0_lblLotProvenance")
	provenance = cleanText(str(provenance))

	# get birth and death date year of artist
	year = re.findall(r'\((.*?)\)',artist_name)
	(birth, death) = ''.join(year).split("-")
	# extract image url from string
	image = re.search("(?P<url>https?://[^\s]+)", str(image)).group("url")

	# extracting width height from details 
	s = cleanText(str(details))
	det= re.findall(r'\((.*?)\)',s)
	for val in det:
	    if 'cm' in val:
	        wt_ht.append(val)
	height, width, length = None, None, None
	a = re.findall(r"[-+]?\d*\.\d+|\d+", ','.join(wt_ht))
	if len(a) == 2:
		height = a[0]
		length = a[1]
	elif len(a) == 3:
		height = a[0] 
		length = a[1]
		width = a[2]        
	elif len(a) == 1:
		height = a[0] 
	artist_name = cleanText(str(artist_name).split('(')[0])

	# append values to a dataframe
	return [ birth, death, art_name, lot_number, artist_name, price, estimate, image, width, height,length, provenance ]
	


if __name__ == '__main__':


	# sample data urls
	pg = links_list

	art_val = pd.DataFrame()

	art_name_l, lot_number_l, artist_name_l, price_l, estimate_l, image_l, width_l, height_l,length_l, provenance_l = [], [] , [] , [] , [], [] , [] , [] , [], []
	birth_l, death_l = [], []

	for pg_val in pg:
		try:
			print(pg_val)
			# data  = scrapeLots(pg_val)
			bd, dt, name , num, art_n, price, est, img, wth, ht, ln , pvnce  = scrapeLots(pg_val)

			# print(name,num,art_n,price,est,img,wth,ht,pvnce)
			art_name_l.append(str(name))
			birth_l.append(bd)
			death_l.append(dt)
			lot_number_l.append(str(num))
			artist_name_l.append(str(art_n))
			price_l.append(str(price))
			estimate_l.append(str(est))
			image_l.append(str(img))
			width_l.append(str(wth))
			height_l.append(str(ht))
			length_l.append(str(ln))
			provenance_l.append(str(pvnce))
		except:
			continue
```

```python
df = pd.DataFrame({'art_name': art_name_l, 'birth_date': birth_l, 'death':death_l, 'lot_number': lot_number_l, 'artist_name': artist_name_l, 'price':price_l, 'estimate':estimate_l, 'image':image_l, 'width': width_l, 'height':height_l,'length': length_l, 'provenance' : provenance_l })
df.head(5)
```

```python
# selecitng columns which are not numeric
alpha = df.columns.drop(['art_name' , 'artist_name', 'price',
       'estimate', 'image','provenance'])
# copying df to new df
sd = df
# converting columns to float
sd[alpha] = df[alpha].apply(pd.to_numeric, errors='coerce')
```

```python
def similarityScore(user_threshold):
	max_threshold = user_threshold + 10
	min_threshold = user_threshold - 10
	omega = sd.loc[(sd['height'] <= max_threshold) & (sd['height'] >= min_threshold)& (sd['length'] >= min_threshold) & (sd['length'] <= max_threshold) ]
	return omega
```
